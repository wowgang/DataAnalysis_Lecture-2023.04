```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('static/img/deadlift.jpg')
plt.imshow(img), plt.axis('off');
```

```python
# 높이,너비,색
np.array(img).shape
```

```python
h, w, _ = np.array(img).shape
```

```python
# 새로운 크기 주기
if h > w:
    width, height = w, w
else:
    width, height = h, h
```

```python
diff = abs(h - w) // 2 # 절대치로 주고 2로 나눈 몫
```

```python
if w > h:   # landscap image
    final_img = np.array(img)[:, diff:diff+h, :]
else:
    final_img = np.array(img)[diff:diff+w, :, :]
```

```python
# 0~ 168 (168, 300, 3)
new_img = Image.fromarray(final_img)
plt.imshow(new_img)
```
