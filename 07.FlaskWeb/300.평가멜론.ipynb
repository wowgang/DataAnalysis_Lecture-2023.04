```python
# 멜론차트 Top100(https://www.melon.com/chart/)을 크롤링을 하여

# 1. 순위, 제목, 아티스트, 앨범, 이미지 필드를 데이터프레임으로 만드세요.[20]

# 2. 위에서 만든 데이터프레임을 멜론_23051211.csv 파일에 저장하세요.[10]   파일이름은 반드시 연월일시를 각각 2자리씩 표시하도록 해야 합니다.

# 3. 웹 서버를 만들고 localhost:5000/melon을 주소창에 입력하면 멜론차트를
#    크롤링한 결과를 표로 보여주는 코드를 작성하고
#    결과 웹화면을 스캔해서 답안지에 포함하세요.[20]
```

```python
import requests
from urllib.parse import quote
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.melon.com/chart/'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
result = requests.get(url,headers=header)
result.text
```

```python
soup = BeautifulSoup(result.text,'html.parser')
```

```python
len(result.text) 
```

```python
trs = soup.select('tr.lst50,tr.lst100')
len(trs)
```

```python
tr = trs[0]
```

```python
# 순위
rank = tr.select_one('.rank').get_text()
rank
```

```python
# 제목
title = tr.select_one('div.ellipsis.rank01 > span > a').get_text().strip()
title
```

```python
# 아티스트
artist = tr.select_one('div.ellipsis.rank02 > a').get_text().strip()
artist
```

```python
# 앨범
album = tr.select_one('.ellipsis.rank03 > a').get_text().strip()
album
```

```python
# 이미지
img = tr.select_one('.image_typeAll > img')['src']
img
```

```python
lines = [] 
for tr in trs:
    rank = tr.select_one('.rank').get_text()
    title = tr.select_one('div.ellipsis.rank01 > span > a').get_text().strip()
    artist = tr.select_one('div.ellipsis.rank02 > a').get_text().strip()
    album = tr.select_one('.ellipsis.rank03 > a').get_text().strip()
    img = tr.select_one('.image_typeAll > img')['src']
    lines.append({ '순위': rank, '제목':title, '아티스트':artist, '앨범':album, '이미지':img})
```

```python
df = pd.DataFrame(lines)
df.tail()
```

```python
# 2. 위에서 만든 데이터프레임을 멜론_23051211.csv 파일에 저장 
# 파일이름은 반드시 연월일시를 각각 2자리씩 표시
```
