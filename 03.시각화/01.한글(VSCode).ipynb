- 라인 플롯

```python
plt.plot([0,1,2,3,],[0,1,4,9]) # X,Y값 주기 /하나만 주면 y값 준것 y=x**엑스제곱
plt.title('Line plot') # 제목
plt.grid() # 격자선
plt.show() # 보여줘라
```
