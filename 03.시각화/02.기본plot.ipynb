### 1. Malplotlib 소개

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # 하이픈 문제?
```

##### 1.라인 플롯

```python
plt.plot([0,1,4,9])     # X가 생략됨. 생략되면 [0,1,2, ...] Y의 개수만큼
plt.title('라인 플롯');plt.grid() # 콤마로 이어도 돼고, 새미콜론으로 이어됨 /한줄에 여러라인쓰기가능
# plt.show() 생략가능
```

##### 1.1 라인 플롯의 스타일 지정

```python
# rs--의 의미는 color, marker, linestyle / r: red ,s:squre 정사각형으로 값 표시, --: 점선(대쉬대쉬)
plt.plot([0,1,4,9], 'rs--')     
plt.title("'rs--' 스타일의 플롯")
plt.grid()
```

```python
plt.plot([0,1,4,9], 'mo-.')     # color, marker, linestyle
plt.plot([9,4,1,0], 'g^:')
plt.title("'mo-.','g^:' 스타일의 플롯")
plt.grid()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'bo--',
         linewidth=5, markersize=15, markeredgecolor="g", markeredgewidth=5, markerfacecolor="r") 
# lw(linewidth) 선 굵기, ms(markersize)마커 크기, mec(markeredgecolor)마커 선 색깔,mew(markeredgewidth),mfc(markerfacecolor)마커 내부 색깔
plt.title("스타일 적용 예"),plt.grid()
plt.show()
```

##### 1.2 그림 범위
    - 범위를 수동으로 지정하려면 xlim 명령과 ylim 명령을 사용한다. 
    - 이 명령들은 그림의 범위가 되는 x축, y축의 최소값과 최대값을 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.grid()
plt.show()
```

- 곡선 그림

```python
X = np.linspace(0, 2*np.pi, 500)    # X = np.arange(0, 2*np.pi+0.01, 0.01) 동일한 코드
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sine, Cosine 그래프')
plt.grid() 
plt.show()
```

##### 1.3 Tick 설정

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]) # 5개의 값
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.plot(X, Ysin), plt.plot(X, Ycos) # plot여러번 쓰면 그래프 겹쳐짐
plt.title('LaTeX, 문자열로 tick label 정의')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # LaTeX 수학 문자식
plt.yticks([-1, 0, 1]), plt.grid()
plt.show()
```

- 범례

```python
plt.plot(X, Ysin, label='Sine')
plt.plot(X, Ycos, label='Cosin') # plot여러번 쓰면 그래프 겹쳐짐
plt.title('범례 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # LaTeX 수학 문자식
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc='upper center') # 범례  # best position = 0 기본값()/ upper center=9 /문자로 줘도 돼고 숫자로 줘도됨. 
plt.show() 
```

- X축, Y축 Label

```python
plt.plot(X, Ysin, label='Sine')
plt.plot(X, Ycos, label='Cosin') # plot여러번 쓰면 그래프 겹쳐짐
plt.title('X축, Y축 Label')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # LaTeX 수학 문자식
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc='upper center') # 범례  # best position = 0 기본값()/ upper center=9 /문자로 줘도 돼고 숫자로 줘도됨. 
plt.xlabel('시간'), plt.ylabel('진폭') # X축, Y축 Label
plt.show() 
```

- 그림의 크기 지정

```python
plt.figure(figsize=(10,6)) # 그림 크기 지정
plt.plot(X, Ysin, label='Sine')
plt.plot(X, Ycos, label='Cosin') # plot여러번 쓰면 그래프 겹쳐짐
plt.title('그림 크기 지정(10,6)')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # LaTeX 수학 문자식
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc='upper center') # 범례  # best position = 0 기본값()/ upper center=9 /문자로 줘도 돼고 숫자로 줘도됨. 
plt.xlabel('시간'), plt.ylabel('진폭')  # X축, Y축 Label
plt.show() 
```

- 그림 저장

```python
plt.figure(figsize=(10,6)) # 그림 크기 지정
plt.plot(X, Ysin, label='Sine') # 범례 label
plt.plot(X, Ycos, label='Cosin') # plot여러번 쓰면 그래프 겹쳐짐
plt.title('그림 저장', fontsize=15) # 폰트 사이즈 지정
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # LaTeX 수학 문자식
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc='upper center') # 범례  # best position = 0 기본값()/ upper center=9 /문자로 줘도 돼고 숫자로 줘도됨. 
plt.xlabel('시간'), plt.ylabel('진폭')  # X축, Y축 Label
plt.savefig('data/삼각함수.png') # 그림저장
plt.show() 
```
