    - Figure는 그림이 그려지는 캔버스나 종이를 뜻하고 Axes는 하나의 플롯, 그리고 Axis는 가로축이나 세로축 등의 축을 뜻한다. Axes와 Axis의 철자에 주의한다.

- Anscombe's Quartet

```python
import seaborn as sns
ans = sns.load_dataset("anscombe")
```

```python
plt.figure(figsize=(10,8))

for index, roman in enumerate(ans.dataset.unique()):  # enumerate 인덱스값 순서를 줌.. enumerate(['I', 'II', 'III', 'IV'])
    ds = ans[ans.dataset == roman]
    plt.subplot(2,2,index+1)    # 여기는 0123아니고 123이라서 index에 +1 / 행의개수,열의개수, 순서(1부터시작)
    plt.scatter(ds.x,ds.y, marker='o') 
    plt.plot([3,20],[4.5,13], 'r-', lw=2) # y = 0.5 * x + 3 회귀식이용 plot이 선......
    plt.title(f'anscombe Dataset_{roman}')
plt.show() # for loop 다 돌고 출력해야해서
```

```python
# y 축 scale 맞춰서 세련되게 다시 그림..
plt.figure(figsize=(10,8))

for index, roman in enumerate(ans.dataset.unique()):  # enumerate 인덱스값 순서를 줌.. enumerate(['I', 'II', 'III', 'IV'])
    ds = ans[ans.dataset == roman]
    plt.subplot(2,2,index+1)    # 여기는 0123아니고 123이라서 index에 +1
    plt.ylim(2,14)
    plt.scatter(ds.x,ds.y, marker='o') 
    plt.plot([3,20],[4.5,13], 'r-', lw=2) # y = 0.5 * x + 3 회귀식이용 plot이 선......
    plt.title(f'anscombe Dataset_{roman}')
plt.show() # for loop 다 돌고 출력해야해서
```

- subplots() 
    - 하나의 그래프에 좌표측이 다른 그래프도 그릴수 있음

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True) # fig는 전체 그림 axes는 그림안의 그림하나

for index, roman in enumerate(ans.dataset.unique()):  # enumerate 인덱스값 순서를 줌.. enumerate(['I', 'II', 'III', 'IV'])
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2] # row index, dolumn index //요고 기억하기...............
    ax.scatter(ds.x,ds.y, marker='o') 
    ax.plot([3,20],[4.5,13], 'r-', lw=2) # y = 0.5 * x + 3 회귀식이용 plot이 선......
    ax.set_title(f'anscombe Dataset_{roman}')

fig.suptitle("Anscombe's Quartet", fontsize=15) # 슈퍼타이틀
plt.tight_layout() # 그림과 그림사이가 더 촘촘하게
plt.show() # for loop 다 돌고 출력해야해서
```

```python
# 다양한 컬러 막대그래프 bar chart
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange'] # color

plt.bar(fruits, counts, label=bar_labels, color=bar_colors) # 모든 ex를 plt로 바꿔줌

plt.ylabel('fruit supply') # set을 지워줌
plt.title('Fruit supply by kind and color') # set을 지워줌
plt.legend(title='Fruit color')

plt.show()
```

- Iris 모든 품종의 특성에 대한 박스플롯을 하나의 그래프안에 그리기

```python
iris = sns.load_dataset('iris')
iris
```

```python
ssl = iris[iris.species == 'setosa'].sepal_length
ssw = iris[iris.species == 'setosa'].sepal_width
plt.figure(figsize=(4,2))
plt.boxplot((ssl,ssw), labels=['sepal_length', 'sepal_width'])
plt.title('Setosa 품종의 꽃받침 길이/너비')
plt.show()
```

```python
# subplot
plt.figure(figsize=(10,8))

for index, species in enumerate(iris.species.unique()): # iris.species.unique()
    sl = iris[iris.species == species].sepal_length
    sw = iris[iris.species == species].sepal_width
    pl = iris[iris.species == species].petal_length
    pw = iris[iris.species == species].petal_width
    plt.subplot(1,3,index+1)    # 여기는 0123아니고 123이라서 index에 +1 / 행의개수,열의개수, 순서(1부터시작)
    plt.boxplot((sl,sw, pl, pw), 
                labels=['s_length', 's_width','p_length','p_width'])
    plt.title(species)
plt.show() # for loop 다 돌고 출력해야해서
```

```python
# subplots
fig, axes = plt.subplots(1, 3, figsize=(10,6), sharey=True)

for index, species in enumerate(iris.species.unique()): 
    sl = iris[iris.species == species].sepal_length
    sw = iris[iris.species == species].sepal_width
    pl = iris[iris.species == species].petal_length
    pw = iris[iris.species == species].petal_width
    ax = axes[index] # 여기
    ax.boxplot((sl,sw, pl, pw), 
                labels=['s_length', 's_width','p_length','p_width'])
    ax.set_title(species)

fig.suptitle("Iris 품종별 특성치의 박스 플롯", fontsize=15) 
plt.tight_layout() 
plt.show() 
```
