### 2. 여러가지 플롯

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

##### 1. Bar Chart

```python
y,x = [2,3,1], [0,1,2]
plt.title('Bar Chart')
plt.bar(x, y)
plt.xticks( x,['가','나','다']) # x의값[0,1,2]에 가,나,다 를 넣어라
plt.yticks(sorted(y))
plt.xlabel('가나다'), plt.ylabel('빈도수')
plt.show()
```

```python
# 판다스 시리즈로 그래프
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar Chart')
plt.yticks([0, 0.2, 0.4, 0.6, 0.8]), plt.xticks([]),plt.grid() # 그리다만 주면 이상해서
plt.show()
```

```python
# 판다스에 플롯이 있음 / 판다스 그래프
# 글자 뒤집어져있음
s.plot(kind='bar')
plt.show()
```

```python
# 판다스 데이터프레임으로 그래프
df = pd.DataFrame(np.random.rand(20).reshape(10,2), 
                  index=list('abcdefghij'), columns=['M','N'])
plt.bar(df.index, df.M.values)
plt.title('Bar Chart')
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
# barh 누워있는 바
s = titanic['class'].value_counts()
plt.figure(figsize=(4,2))
plt.barh(s.index, s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

### 2. Pie Chart

```python
ratio = [10, 15, 20, 25] #합이 100이 아니어도 됨
labels = ['감','귤','배','밤']
plt.pie(ratio, labels=labels , autopct='%.1f%%') # 소숫점한자리까지 %% 맥프로우?
plt.title('파이 차트')
plt.show()
```

```python
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # 두번째 값이 밖으로 튀어 나오게해줘라
plt.pie(ratio, labels=labels , autopct='%.1f%%', colors=colors, explode=explode,
         startangle=90, shadow=True) # 시작위치 90도 
plt.title('파이 차트')
plt.show()
```

##### 4. 산점도(Scatter plot)

```python
np.random.seed(2023)
X,Y = np.random.normal(2, 2,100),np.random.rand(100) # normal 평균2 표준편차2 100개
plt.figure(figsize=(5,3))
plt.scatter(X,Y)
plt.title('산점도')
plt.show()
```

```python
# 산점도에 색 주기
colors = np.random.rand(100) # 0과 1 사이에서 100개의 무작위 수를 생성하여 배열로 반환
# 0에서 1사이인데 색이 나오는 이유는 palette에서 색상을 고르기 때문
np.random.seed(2023)
X,Y = np.random.normal(2, 2,100),np.random.rand(100) # normal 평균2 표준편차2 100개
plt.figure(figsize=(5,3))
plt.scatter(X,Y, c=colors, alpha=0.5) # 색,투명도
plt.title('색상을 준 산점도')
plt.show()
```

```python
# 버블차트
size = 500 * np.random.rand(100) # 크기
plt.figure(figsize=(5,3))
plt.scatter(X,Y, s=size ,c=colors ,alpha=0.5) # 점사이즈,색,투명도
plt.title('Bubble Chart')
plt.show()
```
