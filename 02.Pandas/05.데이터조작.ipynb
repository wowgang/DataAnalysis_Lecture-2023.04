- 연습 문제 4.4.4

```python
''' 
타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라.
category1 카테고리는 다음과 같이 정의된다.
# 20살이 넘으면 성별을 그대로 사용한다.
# 20살 미만이면 성별에 관계없이 “child”라고 한다.
'''
```

```python
titanic.head()
```

```python
titanic[['sex','age']]
```

```python
titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
# x.sex 값과  x.age 을 동시에 사용해야 하기 때문에 , axis=1을 써 주어야 한개의 열로 가지고 올수 있습니다.
# titanic.apply 로 titanic 전체에 대한 apply이기 때문에 axis=1이 필요합니다
''''apply' 메서드의 'axis=1' 인수는 pandas에게 데이터 프레임의 각 행에 람다 함수를 적용하도록 지시합니다. 
이 작업의 결과는 원래 '타이타닉' 데이터 프레임에 있는 
각 사람의 나이와 성별에 따라 '남성', '여성' 또는 '어린이' 값이 있는 새 열입니다.'''
# 지금 열이 필요한게 아니라 각 행이 필요한것임***axis=1을 써 주어야 한개의 열
```

```python
category1 = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
df = pd.DataFrame({'category1':category1})
df
```
