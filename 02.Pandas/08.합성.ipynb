```python
import numpy as np
import pandas as pd
```

- merge 함수

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
# key 의 list형태로 
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
교집합 inner join
합집합 full outer join
left outer join
right outer join
```

```python
# Inner join 공통된것만 합치기 
# 이름이 같은 열의 이름이 같은것(열 이름이 같은열 -컬럼명이 같은것 /고객번호) 기준으로, 공통된 데이터만 합성
# 컬럼명이 같은 열(고객번호) 기준으로
pd.merge(df1, df2)
```

```python
# Full outer join 전체
pd.merge(df1,df2, how='outer')
```

```python
# Left outer join / 왼df1 , 오df2
pd.merge(df1, df2, how='left')
```

```python
# right outer join / 왼df1 , 오df2
pd.merge(df1, df2, how='right')
```

- join method

```python
# 인덱스로 조인
df1.set_index('고객번호', inplace=True)
df2.set_index('고객번호', inplace=True)
```

```python
df1.join(df2)   # = left outer join
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
'''
두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다.

각각 5 x 5 이상의 크기를 가진다.
공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다
'''
```

```python
df1 = pd.DataFrame(
    np.random.randint(40, 101, 25).reshape(5,5), # 2차원 데이터
    columns=list('ABCDE')  # 컬럼명 list로 줘야함
)
df1
```

```python
df2 = pd.DataFrame(
    np.random.randint(40, 101, 25).reshape(5,5), # 2차원 데이터
    columns=list('ABXYZ')   # 컬럼명 list로 줘야함

)
df2
```

```python
df1.A[0] = 98
df1.B[1] = 70
df1
```

```python
pd.merge(df1,df2, on='A') # A를 기준으로 합친
```

```python
pd.merge(df1,df2, on='B') # B를 기준으로 합친
# df1과 df2의 공통된 값인 B 열에 대해 매칭되는 행만 결과에 포함
```

```python
pd.merge(df1, df2, on=['A', 'B'])
```

- concat 함수를 사용한 데이터 연결
    - concat 함수를 사용하면 기준 열(key column)을 사용하지 않고 단순히 데이터를 연결(concatenate)한다.

    - 기본적으로는 위/아래로 데이터 행을 연결한다. 단순히 두 시리즈나 데이터프레임을 연결하기 때문에 인덱스 값이 중복될 수 있다.

```python
'''
어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 
후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 
실적 정보는 “매출”, “비용”, “이익” 으로 이루어진다. (이익 = 매출 - 비용).

또한 1년간의 총 실적을 마지막 행으로 덧붙인다.
'''
```

```python
df5 = pd.DataFrame(np.random.randint(500,1000, size=(6,2)) * 1000, columns=['매출', '비용'] , index=list('123456'))
df5['이익'] = df5['매출'] - df5['비용']
df5
```

```python
df6 = pd.DataFrame(np.random.randint(500,1000, size=(6,2)) * 1000, columns=['매출', '비용'] , index=['7','8','9','10','11','12'])
df6['이익'] = df6['매출'] - df6['비용']
df6
```

```python
df7 = pd.concat([df5,df6])
df7
```

```python
df7.sum()
```

```python
# 행 추가
df7.loc['합계', :] = df7.sum() # axis=0이 기본값 
df7.tail()  # df7.loc['합계', : 이렇게 슬라이싱하면 data가 실수가 되는데 슬라이싱안하면 정수로됨
```

```python
df7.drop(['합계'], inplace=True) # data실수 값 지우려고 다시할거
```

```python
# type int로 변환후 계산하기
df7 = df7.astype(int)
df7.tail()
```

```python
# 행 추가
df7.loc['합계'] = df7.sum() # axis=0이 기본값 
df7.tail()
```
