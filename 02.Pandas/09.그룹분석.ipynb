##### 1. pivot table
- 피봇테이블(pivot table)이란 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것을 말한다.

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
#  pivot은 중복된 index와 column을 처리하지 못하는데, pivot_table은 처리할 수 있습니다.
```

```python
''' 
1. 기본적인 연산, 평균, 분산, 표준편차, 최댓값, 최솟값, 중앙값 등의 값을 연산하기 위한 목적으로 사용할거라면 
사실 groupby를 쓰든 pivot_table을 쓰든 큰 상관이 없습니다.

다만 groupby 가 좀 더 속도가 빠르고 반환값이 pivot_table은 항상 데이터프레임으로 
groupby는 컬럼이 series 형태라면 series 로 반환된다는게 가장 큰 차이점으로 보일거에요.

2. pivot 은 데이터의 형태를 바꿀 때 사용해요.
그래서 pivot은 연산이 안 되는데 pivot_table 은 연산을 할 수 있어요.
그래서 pivot과 pivot_table의 차이는 연산을 할 수 있느냐 없느냐에요.
'''
```

```python
df1.pivot(values='인구' ,index='도시', columns='연도')
```

```python
# pivot만들기1 data , index columns 순으로
df1.pivot_table(values='인구' ,index='도시', columns='연도')
```

```python
# pivot만들기2 data , index columns 순으로
df1.pivot_table('인구','도시','연도')
```

```python
# 이중으로 인덱스 여러게 주고싶을때
df1.pivot_table('인구',['지역','도시'],'연도') 
```

##### 2. 그룹분석

- groupby 메서드
- 분석하고자 하는 시리즈나 데이터프레임에 groupby 메서드를 호출하여 그룹화를 한다.
- 그룹 객체에 대해 그룹연산을 수행한다.
- groupby 메서드는 데이터를 그룹 별로 분류하는 역할을 한다. groupby 메서드의 인수로는 다음과 같은 값을 사용한다.
    - 열 또는 열의 리스트
    - 행 인덱스

```python
import seaborn as sns
iris = sns.load_dataset('iris') # 붓꽃
tips = sns.load_dataset('tips')
```

##### 2.1 groupby

- iris data

```python
'''
붓꽃(iris) 데이터셋
sepal 꽃받침
petal 꽃잎
species 품종 (setosa,versicolor,virginica)
'''
```

```python
iris.shape # 150행 5열
```

```python
iris.head(3)
```

```python
iris[50:53]
```

```python
iris.tail(3)
```

```python
# 품종별 항목의 개수
iris.groupby('species').count() # 각각 50개씩 data가 있다.
```

```python
# 품종별 항목의 평균
iris.groupby('species').mean() # 평균구하기
```

```python
# 품종별 sepal_length(꽃받침 길이)의 평균 - 결과 series
iris.groupby('species').mean()['sepal_length'] # dataframe에서 하나의 컬럼을 설정하면 시리즈로 반환
```

```python
iris.groupby('species')['sepal_length'].mean()
```

```python
# groupby 해서 dataframe으로
iris[['sepal_length', 'species']].groupby('species').mean()
```

```python
# 품종별 항목의 평균과 표준편차 
# 2개이상을 알고 싶을 때
iris.groupby('species').agg(['mean', 'std'])
# agg메서드 사용하기 agg하고 ([]) 리스트 형태로 넣어주기
```

```python
# 개별 항목의 다양한 통계
iris.groupby('species')['sepal_width'].agg(['mean', 'std', 'min', 'max'])
```

- tips data

```python
print(tips.shape)
tips.head()
```

```python
tips['tip_pct'] = (tips.tip / tips.total_bill *100).round(2)
tips.tail()
```

```python
# 기초통계
tips.describe()
```

```python
# 성별, 흡연유무별 팁 비율의 평균 - series
tips.groupby(['sex','smoker'])['tip_pct'].mean()
```

```python
# 성별, 흡연유무별 팁 비율의 평균 - dataframe
tips.groupby(['sex','smoker'])[['tip_pct']].mean()
```

```python
# 성별, 흡연유무별 팁 비율의 데이터 수, 평균, 최소, 최대
tips.groupby(['sex','smoker'])[['tip_pct']].agg(['count','mean','min','max'])
```

```python
# 성별, 흡연유무별, size별  팁 비율의 데이터 수, 평균, 최소, 최대
tips.groupby(['sex','smoker','size'])[['tip_pct']].agg(['count','mean','min','max'])
```

```python
# 요일별, 시간대별 팁 비율의 데이터 수, 평균, 최소, 최대
tips.groupby(['day','time'])[['tip_pct']].agg(['count','mean','min','max'])
```

##### 2.1 pivot_table

```python
# 성별, 흡연유무별 팁의 비율
tips.pivot_table('tip_pct','sex','smoker', 'mean') # values, index, columns, aggfunc
```

```python
tips.pivot_table('tip_pct','sex','smoker','max')
```

```python
# pivot_table의 합계기능
# 성별, 흡연유무별 인원
tips.pivot_table('tip_pct','sex','smoker', 'count', margins=True, margins_name='계') #  margins_name='계' 안하면 All로 나옴
```

##### 2.3 groupby vs pivot_table

```python
# 성별 팁 비율의 평균groupby
tips.groupby(['sex'])[['tip_pct']].agg(['mean'])
```

```python
# 성별 팁 비율의 평균groupby
tips.groupby('sex')[['tip_pct']].mean()
```

```python
# 성별 팁 비율의 평균 pivot_table
tips.pivot_table('tip_pct','sex') # mean평균이 defalt
```

```python
# 성별 팁 비율의 평균 pivot_table
tips.pivot_table('tip_pct','sex', aggfunc='mean')
```

```python
# 흡연 유무에 따른 평균 팁 비율과 최소값, 최대값
tips.groupby(['smoker'])[['tip_pct']].agg(['mean','min','max'])
```

```python
tips.groupby('smoker')['tip_pct'].agg(['mean','min','max'])
```

```python
tips.pivot_table('tip_pct','smoker', aggfunc=['mean','min','max']) 
```

- 연습문제 4.7.4

```python
df = sns.load_dataset('titanic')
df.head(3)
```

```python
# 1.qcut 명령으로 세 개의 나이 그룹을 만든다.
df['cat'] = pd.qcut(df.age, 3, labels=['A1','A2','A3'])
df.head()
```

```python
df.cat.isna().sum()
```

```python
# 2.성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 
# 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. 
# 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.
# 생존률은 survied을 평균하면 나온다.
df.pivot_table('survived',['sex','cat'], 'class')
```

```python
# 3.성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.
df.pivot_table('survived','sex','class') # mean안써도 평균나오니까
```
