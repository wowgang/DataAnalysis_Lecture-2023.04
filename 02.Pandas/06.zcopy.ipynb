```python
import csv

# txt 파일에서 데이터 추출하기
data = []
with open('data/서울시인구.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        data.append(line.split('\t'))

# 추출한 데이터를 csv 파일에 저장하기
with open('data/서울시인구.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
