```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
'''
1. 전국도시공원정보 표준데이터 파일을 읽고 다음에 답하세요. [40]
1) 공원명, 소재지지번주소, 공원면적 컬럼만 선택해서 데이터 프레임을 만드세요.
2) 경기도 데이터만 추출하세요.
3) 소재지지번주소에서 시군구(예, 이천시, 양평군, 수원시 장안구)를 추출해서   시군구 컬럼을 만드세요.
4) 시군구 별로 공원면적의 합을 구하세요.
'''
```

```python
# 1) 공원명, 소재지지번주소, 공원면적 컬럼만 선택해서 데이터 프레임을 만드세요.
gpark = pd.read_csv('../04.지도시각화/data/경기도도시공원정보표준데이터.csv', encoding='euc-kr')
gpark.info()
```

```python
gp = gpark[['공원명','소재지지번주소' ,'공원면적']]
gp.head(110).tail(10)
```

```python
# 2) 경기도 데이터만 추출하세요.
gp.dropna(subset=['소재지지번주소'], inplace=True)
gp.isna().sum() 
```

```python
# 2) 경기도 데이터만 추출하세요.
gdf = gp[gp.소재지지번주소.str.contains('경기도')]
gdf.head(80).tail(10)
```

```python
# 3) 소재지지번주소에서 시군구(예, 이천시, 양평군, 수원시 장안구)를 추출해서   시군구 컬럼을 만드세요.
gdf['시군구'] = gdf.소재지지번주소.apply(lambda x: ''.join(x.split()[1:3]) if x.find('구 ') >= 0 else x.split()[1])
gdf
```

```python
type(gp.공원면적[0])
```

```python
gdf.공원면적 = gdf.공원면적.astype(int)
```

```python
# 4) 시군구 별로 공원면적의 합을 구하세요.
gdf1 = gdf.groupby(['시군구'])[['공원면적']].agg(['sum'])
gdf1.head()
```

```python
'''
1. 경기도 주민등록인구집계현황 파일을 읽고 다음에 답하세요. [40]
1) 행정구역구분명, 행정구역명, 총 인구수 컬럼만 선택해서 데이터프레임을   만드세요.
2) 경기도와 읍면동 인구를 제외한 시군과 구의 데이터만 추출하세요.   
(단, 양주시와 광주시 오포2동은 데이터를 고쳐서 사용하세요.) 
3) 수원시, 용인시 등 구가 있는 시는 시 데이터를 제거하세요.
4) 시군구 컬럼을 만드세요.
'''
```

```python
# 1) 행정구역구분명, 행정구역명, 총 인구수 컬럼만 선택해서 데이터프레임을   만드세요.
pop = pd.read_csv('../02.Pandas/data/주민등록인구집계현황.csv', encoding='euc-kr')
pop = pop[['행정구역구분명','행정구역명','총 인구수']]
pop.head()
```

```python
# 2) 경기도와 읍면동 인구를 제외한 시군과 구의 데이터만 추출하세요. 
# (단, 양주시와 광주시 오포2동은 데이터를 고쳐서 사용하세요.) 
```

```python
except_city = ['경기도 수원시', '경기도 고양시', '경기도 용인시', '경기도 성남시', '경기도 안산시', '경기도 안양시']

pop = pop[((pop.행정구역구분명 == '시군') | (pop.행정구역구분명 == '구')) & ~(pop['행정구역명'].str.strip().isin(except_city))]
pop
```

```python
pop['시군구'] = pop.행정구역명.apply(lambda x: ''.join(x.split()[1:3]) if x.find('구 ') >= 0 else x.split()[1])
pop
```
