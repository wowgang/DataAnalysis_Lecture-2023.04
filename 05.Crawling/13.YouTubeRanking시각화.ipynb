```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/20230425_YouTube랭킹.csv')
df.head(3)
```

- 구독자수 Top 20 - barh 차트로
- 조회수 Top 20 - barh 차트로
- 비디오수 Top 20 - barh 차트로
- 카테고리별 채널수 Top 10
- 카테고리별 구독자수 Top 10

```python
plt.figure(figsize=(8,4))
plt.barh(df['채널명'][:20], df['구독자수'][:20])
plt.title('구독자수 Top 20 - barh')
plt.show()
```

```python
# 선생님방법
plt.figure(figsize=(8,4))
plt.barh(df.채널명.head(20), df.구독자수.head(20))
plt.title('구독자수 Top 20 - barh')
plt.show()
```

```python
# 선생님sns
plt.figure(figsize=(8,4))
sns.barplot(y='채널명', x='구독자수', data=df.head(20))
plt.title('구독자수 Top 20 - barh')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(df['채널명'][:20], df['조회수'][:20])
plt.title('조회수 Top 20 - barh')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(df['채널명'][:20], df['비디오'][:20])
plt.title('비디오수 Top 20 - barh')
plt.show()
```

```python
df2 = df.sort_values(by='비디오', ascending=False)

plt.figure(figsize=(8,4))
sns.barplot(y='채널명', x='비디오', data=df.head(20))
plt.title('비디오 Top 20')
plt.show()
```

```python
# .to_frame() 시리지를 바로 프레임으로 바꿔줘라
df4 = df['카테고리'].value_counts().to_frame()
df4
```

```python
df.pivot_table('채널명', '카테고리', aggfunc='count').sort_values('채널명', ascending=False)
```

```python
ca = df['카테고리'].value_counts().head(10)
caa = pd.DataFrame(ca)
caa
```

```python
caa.index
```

```python
plt.figure(figsize=(8,4))
plt.barh(caa.index, caa['count'])
plt.title('카테코리별 채널수 Top 10 - barh')
plt.show()
```

```python
sns.barplot(y=df4.index[:10], x='count', data=df4.head(10));
plt.title('카테코리별 채널수 Top 10')
```

```python
plt.figure(figsize=(8,4))
plt.barh(df['카테고리'], df['구독자수'])
plt.title('카테코리별 구독자수 Top 10 - barh')
plt.show()
```

```python
# data 인덱스
df5 = df.pivot_table('구독자수','카테고리', aggfunc='sum' ).sort_values('구독자수', ascending=False)
df5
```

```python
sns.barplot(y=df5.index[:10], x='구독자수', data=df5.head(10));
plt.title('카테코리별 채널수 Top 10')
```

```python
sns.barplot(y=df5.index[:10], x=df5.구독자수.head(10)) # data안줘도 구해짐
plt.title('카테코리별 채널수 Top 10');
```
