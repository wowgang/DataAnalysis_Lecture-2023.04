##### 4. 한 페이지내에 있는 정보

```python
# 방법1. 딕셔너리의 리스트로 가져오기 
line = []
# trs = soup.select('#body-content > div.newest-list > div > table > tbody > tr')
for tr in trs:
    rank = tr.select_one('.number').get_text().split('\n')[0].strip()
    img = 'https:' + tr.select_one('.cover > img')['src']
    try:
        title = tr.select_one('.title.ellipsis').get_text().strip() #19금아닌것
    except:
        title = tr.select_one('.title.ellipsis').get_text().split('\n')[-1].strip() # 19금
    artist = tr.select_one('.artist.ellipsis').string.strip()
    album = tr.select_one('.albumtitle.ellipsis').text.strip() 
    line.append({'rank':rank, 'img':img, 'title':title, 'artist':artist, 'album':album})
```

```python
df1 = pd.DataFrame(line)
df1.head()
```

```python
# 방법 2. 개별 리스트로 가져오기
rank_list, img_list, title_list, artist_list, album_list = [], [], [], [], []
# trs = soup.select('#body-content > div.newest-list > div > table > tbody > tr')
for tr in trs:
    rank_list.append(tr.select_one('.number').get_text().split('\n')[0].strip())
    img_list.append('https:' + tr.select_one('.cover > img')['src'])
    try:
        title_list.append(tr.select_one('.title.ellipsis').get_text().strip()) #19금아닌것
    except:
        title_list.append(tr.select_one('.title.ellipsis').get_text().split('\n')[-1].strip()) # 19금
    artist_list.append(tr.select_one('.artist.ellipsis').string.strip())
    album_list.append(tr.select_one('.albumtitle.ellipsis').text.strip())
```

```python
df2 = pd.DataFrame({
    'rank': rank_list, 'img':img_list, 'title':title_list, 'artist':artist_list, 'album':album_list
})
df2.head()
```

##### 6. 모든 페이지에 대해서 처리하기

```python
# 페이지를2,3, 다녀온뒤 1로 다시 돌아오니
# url 형식이 달라짐
# pg=1
```

```python
# 실시간차트
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230424&hh=15&rtm=Y&pg=1
from datetime import datetime
now = datetime.now()
now
```

```python
# ymd=20230424
ymd = now.strftime('%Y%m%d') 
# 방법2 ymd = f'{now.year}{now.month:02d}{now.day:02d}'
# hh=15
hh = now.strftime('%H')
# # 방법2  hh = f'{now.hour:02d}'
ymd, hh
```

```python
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230424&hh=15&rtm=Y&pg=1
# https://www.genie.co.kr/chart/top200?ditc=D
# pg=1
# &rtm=Y
base_url = 'https://www.genie.co.kr/chart/top200?ditc=D&rtm=Y'
```

```python
from tqdm import tqdm # 진척률
lines = []
base_url = 'https://www.genie.co.kr/chart/top200?ditc=D&rtm=Y'
for page in tqdm(range(1, 5)): # 진척률 4페이지 반복
    url = f'{base_url}&ymd={ymd}&hh={hh}&pg={page}'
    result = requests.get(url, headers=header) # header 내가 로봇이 아니라는걸...말해주는?
    soup = BeautifulSoup(result.text, 'html.parser')
    trs = soup.select('tr.list')
    for tr in trs:
        rank = tr.select_one('.number').get_text().split('\n')[0].strip()
        img = 'https:' + tr.select_one('.cover > img')['src']
        try:
            title = tr.select_one('.title.ellipsis').get_text().strip() #19금아닌것
        except:
            title = tr.select_one('.title.ellipsis').get_text().split('\n')[-1].strip() # 19금
        artist = tr.select_one('.artist.ellipsis').string.strip()
        album = tr.select_one('.albumtitle.ellipsis').text.strip() 
        lines.append({'rank':rank, 'img':img, 'title':title, 'artist':artist, 'album':album})
```

##### 7. 데이터프레임 만들기

```python
df = pd.DataFrame(lines)
df
```

```python
df.to_csv(f'data/{ymd}{hh}_지니차트.csv', index=False)
```
