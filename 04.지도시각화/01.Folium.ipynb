### folium 지도 시각화

- 설치 : pip install folium

```python
import folium
folium.__version__
```

```python
'''
folium.Map()을 이용하여 처음 화면이 나오는 부분을 설정 # open street map (default)
location은 지도의 기준이 될 위도(lat)와 경도(lon)의 좌표를 전달 # 센터를 의미
zoom_start는 처음 지도를 봤을 때의 확대 정도를 나타내고 최댓값은 18
만약 location 값과 zoom_start 값을 지정하지 않으면 전 세계 지도를 기준으로 띄어줍니다

기본형 => folium.Map(location=[위도, 경도], zoom_start=숫자)

folium.Marker 지도에 표시
'''
```

```python
map = folium.Map(location=[37.295, 127.045], zoom_start=14) # tiles=open street map (default)
folium.Marker(
    location=[37.2949837, 127.0453902], # WGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200), # 마커클릭시 뜨는 문구를 popup
    tooltip='경기도청' # 마우스를 마커위에 올려놨을때 뜨는 문구
).add_to(map)
map
```

```python
# Stamen Toner Map
map = folium.Map(location=[37.295, 127.045], zoom_start=14, tiles='Stamen Toner')
folium.Marker(
    location=[37.2949837, 127.0453902], # WGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200), # 마커클릭시 뜨는 문구를 popup
    tooltip='경기도청' # 마우스를 마커위에 올려놨을때 뜨는 문구
).add_to(map)
map
```

```python
map = folium.Map(location=[37.285, 127.005], zoom_start=12) # tiles=open street map (default)
folium.Marker(
    location=[37.2949837, 127.0453902], # WGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200), # 마커클릭시 뜨는 문구를 popup
    tooltip='경기도청' # 마우스를 마커위에 올려놨을때 뜨는 문구
).add_to(map)
folium.Marker(
    location=[37.30337225, 127.0096118], 
    popup=folium.Popup('경기도 수원시 장안구 송원로 101', max_width=200),
    tooltip='장안구청', 
    icon=folium.Icon(
    color='purple', icon_color='white',
    icon='glyphicon glyphicon-thumbs-up', # 아이콘 제공 사이트
    prefix='glyphicon')
).add_to(map)
folium.Marker(
    location=[37.28256171, 127.0190196],
    popup=folium.Popup('경기도 수원시 팔달구 창룡대로 23', max_width=200), 
    tooltip='팔달구청', 
    icon=folium.Icon(icon='cloud' , color="red")
).add_to(map)
folium.Circle(
    radius=300,color="crimson",fill=False,
    location=[37.25764489, 126.9719386], 
    popup=folium.Popup('경기도 수원시 권선구 호매실로 12', max_width=200), 
    tooltip='권선구청' 
).add_to(map)
folium.CircleMarker(
    radius=15,color="#3186cc",fill=True,fill_color="#3186cc",
    location=[37.2596127, 127.0465469], 
    popup=folium.Popup('경기도 수원시 영통구 효원로 407', max_width=200), 
    tooltip='영통구청'
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">수원 소재 주요 관공서</h3>' # 지도 제목달기
map.get_root().html.add_child(folium.Element(title_html))
map
```
