### 단계 구분도(Choropleth)

```python
import json, folium
import pandas as pd
```

- 서울시 구별 CCTV 설치대수

```python
# 단계 구분도(Choropleth maps) 하려면 simple.json파일 다운받아야함
```

- 데이터 준비

```python
df = pd.read_csv('../02.Pandas/data/cctv1.csv', index_col='구별')
df.head(3)
```

```python
# 데이터프레임의 구별과 서울제이슨파일의 id와 매치
```

```python
file = 'data/seoul_geo_simple.json'
geo_data = json.load(open(file, encoding='utf-8'))
```

- 단계구분도

```python
map = folium.Map([37.55, 126.98], zoom_start=11, tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data, # 맵데이터
    data=df.CCTV갯수, # 단계구분도로 보여줄 데이터
    columns=[df.index, df.CCTV갯수], # dataframe에서 추출할 컬럼(항목)
    fill_color='YlOrRd',   # Colormap
    key_on='feature.id' # 지도에서 join할 항목
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">서울시 자치구별 CCTV설치갯수</h3>' # 지도 제목달기
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
# 더 보기좋은 지도 한글로 경도위도를 한글로 표시하기위한 함수 선생님
# 함수선언
import numpy as np
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']: # 경도,위도 표시
            geo = np.array(coord)
            gu_dict[gu['id']] = [np.mean(geo[:,1]), np.mean(geo[:,0])]  # 경도,위도 표시
    return gu_dict
```

```python
gu_dict = get_text_location(geo_data) # /seoul_geo_simple
gu_dict
```

```python
# 
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
                              html=f'<div style="font-size: 10pt">{gu_name}</div>')
    ).add_to(map)
map
```
